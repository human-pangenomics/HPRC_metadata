```python
import pandas as pd
```

```python
hifi_qc = pd.read_csv('/private/groups/hprc/human-pangenomics/HPRC_metadata/submissions/HPRC_Primrose_Y1-Y3-redo/hifi_qc/hifi_qc_samples_output_table.csv')
```
