```python
import pandas as pd
```

```python
hifi_qc = pd.read_csv('/private/groups/hprc/human-pangenomics/HPRC_metadata/submissions/HPRC_Primrose_Y1-Y3/hifi_qc/hifi_qc_samples_output_table.csv')
```

```python
hifi_dict = {sample_data[0]:pd.read_csv(sample_data[1],sep='\t') for sample_data in hifi_qc[['sample_id', 'hifi_qc_summary']].values}
```

```python
hifi_df = pd.concat(hifi_dict.values())
```

```python
hifi_df.drop(columns=['all_kinetics_flag', 'keep_kinetics_flag', 'hifi_kinetics_tag', 'PP_PRIMROSE', 'fi_tag', 'ri_tag','fp_tag', 'rp_tag'],inplace=True)
```

```python
hifi_df.sort_values(by=['ntsm_score'],ascending=False).to_csv('HPRC_Primrose_Y1-Y3_sample_file_hifi_qc.csv',index=False)
```
