```python
def aws_working(file_paths):
	new_path_list = []
	for data in file_paths:
		sample_id = data.split('/')[-4]
		print(sample_id)
		print(data)
		new_path = f"s3://human-pangenomics/working/HPRC/{sample_id}/raw_data/hic/{data.split('/')[-1]}"
		new_path_list.append(new_path)
	return new_path_list
hic_y3_working = aws_working(hic_y3['hic_reads'].tolist())
```

```python
aws_df = pd.DataFrame(hic_y3['hic_reads'])
aws_df['release'] = hic_y3_working
```
