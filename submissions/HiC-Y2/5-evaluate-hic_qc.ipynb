```python
import pandas as pd
```

```python
hic_output_df = pd.read_csv('hic_qc/hic_samples_output_table.csv')
```

```python
# hic_output_df[hic_output_df['sample_id'].isin(['HG02071-10a-1_S9_L002_R2_001.fastq.gz'])]
```

```python
hic_list = []
for qc in hic_output_df[['sample_id', 'hic_qc_summary']].values:
	print(qc[0])
	hic_temp = pd.read_csv(qc[1],sep='\t')
	hic_list.append(hic_temp)
hic_qc = pd.concat(hic_list)
```

```python
hic_qc['sample'] = [sample_id.split('-')[0] for sample_id in hic_qc['file_name'].tolist()]
```

```python
hic_qc.shape
```

```python
hic_qc[hic_qc['ntsm_score'].apply(lambda x: x > 0.9 or x < 0.3)].to_csv('HIC_Y2_sample_file_hic_qc_removal.csv',index=False)
```
