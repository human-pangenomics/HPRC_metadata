```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
sra_ids = pd.read_csv('/private/groups/hprc/human-pangenomics/documentation-metadata/HPRC_metadata/submissions/HiC-Y2/metadata-14716579-processed-ok.tsv',sep='\t')

release = pd.read_csv('/private/groups/hprc/human-pangenomics/documentation-metadata/HPRC_metadata/submissions/HiC-Y2/HiC-Y2.transfer.csv',header=None)
release['filename'] = [filename.split('/')[-1] for filename in release[1].tolist()]
release.rename(columns={1:'path'},inplace=True)
```

```python
hic_file_df = pd.DataFrame(sra_ids['filename'].tolist() + sra_ids['filename2'].tolist())

# Create a new DataFrame with the rows duplicated for filename2
df_filename2 = sra_ids.copy()
df_filename2['filename'] = df_filename2['filename2']

# Combine the original and the new DataFrame, then drop filename2 column
df_combined = pd.concat([sra_ids, df_filename2], ignore_index=True)
df_combined = df_combined.drop(columns=['filename2'])

# Display the resulting DataFrame
# print(df_combined)
```

```python
hic_qc = pd.read_csv('/private/groups/hprc/human-pangenomics/documentation-metadata/HPRC_metadata/submissions/HiC-Y2/HIC_Y2_sample_file_hic_qc.csv')
hic_qc.rename(columns={'file_name':'filename'},inplace=True)
```

```python
hic_df = pd.merge(df_combined, hic_qc, on='filename',how='inner')
```

```python
hic_df['assembly'] = 'unaligned'
```

```python
hic_df = pd.merge(hic_df, release[['path','filename']], on='filename', how='inner')
hic_df['total_Gbp'] = round(hic_df['total_bp'] / 1_000_000_000,2)
hic_df['sample_id'] = [sample.split('-')[0] for sample in hic_df['library_ID'].tolist()]
hic_df.drop(columns=['object_status'],inplace=True)
```

```python
hic_df.to_csv('HIC_Y2_final_table.csv',index=False)
```
